```python
# Creates the giant graph

g = nx.Graph()
actors_dict = {}
movie_rating_dict = {}
movie_budget_dict = {}
movie_revenue_dict = {}
movie_profit_dict = {}
movie_cast_dict = {}

for m in movie_info:
    cast_list = m[-1]
    try:
        cast_list.remove('')
    except ValueError:
        pass
    movie_rating_dict[m[1]] = float(m[3])
    movie_budget_dict[m[1]] = float(m[4])
    movie_revenue_dict[m[1]] = float(m[5])
    movie_profit_dict[m[1]] = float(m[6])
    movie_cast_dict[m[1]] = cast_list
    for i in range(0,len(cast_list)):
        try:
            actors_dict[cast_list[i]] += (','+m[1])
        except KeyError:
            actors_dict[cast_list[i]] = m[1]
        for j in range(i+1, len(cast_list)):
            new_edge = None
            try:
                g[cast_list[i]][cast_list[j]]['weight'] += 1
            except KeyError, TypeError:
                g.add_edge(cast_list[i], cast_list[j], weight=1)

print 'Nodes: ' + str(len(g.nodes()))
print 'Edges: ' + str(len(g.edges()))
```

```python
# Removes edges of weight 1

edge_remove = []
for edge in g.edges(data=True):
    if edge[2]['weight'] < 2:
        edge_remove.append(edge)
g.remove_edges_from(edge_remove)

node_remove = [node for node,degree in g.degree().items() if degree == 0]
g.remove_nodes_from(node_remove)

print 'Nodes: ' + str(len(g.nodes()))
print 'Edges: ' + str(len(g.edges()))
```

```python
# Functions for investigation and such

# O(V^2 log(V))
def max_weight(g):
    return max(g.edges(data=True), key=lambda (source,target,data): data['weight'])

# O(V)
def max_degree(g):
    return max(nx.degree(g).values())

# O(V)
def node_of_max_degree(g):
    max_d = max_degree(g)
    node = [n for n in g if g.degree(n) == max_d]
    return g.degree(node)


def top_x_percentile_degrees(g, x):
    p = 1-(x*0.01)
    top = []
    max_d = max_degree(g)
    node = [n for n in g if g.degree(n) >= max_d*p]
    return g.degree(node)

# dictionary sorting by values taken from http://stackoverflow.com/questions/613183/sort-a-python-dictionary-by-value
def top_x_degrees(g, x):
    sorted_d = sorted(nx.degree(g).items(), key=operator.itemgetter(1), reverse=True)
    if x < len(sorted_d):
        return sorted_d[:x+1]
    else:
        return sorted_d
    
print top_x_degrees(g, 25)
```

```python
L = nx.laplacian_matrix(cc_g).astype(float)
w,v = sp.sparse.linalg.eigsh(L, k = 3, which='SM')
```

```python
m = 500

X=v*w

kmeans = KMeans(init='k-means++', n_clusters=m, n_init=10)
kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_
```

```python
# create communities and compute averages

cluster = [[] for i in range(0, m)]

nodes_list = cc_g.nodes()
for i in range(0,len(labels)):
    idx = labels[i]
    cluster[idx].append(nodes_list[i])

total = 0
big_group = []
for i in cluster:
    total += len(i)
    big_group += i

print total

com_movies = [[] for i in range(0, m)]

mismatches = 0
for i in range(0, m):
    movies = []
    people = cluster[i]
    if len(people) < 5:
        com_movies[i] = []
    else:
        for p in people:
            try:
                movies += actors_dict[p].split(',')
            except KeyError:
                mismatches +=1
        movies = list(set(movies))
        for mov in movies:
            cast_list = movie_cast_dict[mov]
            count = 0
            for p in people:
                if p in cast_list:
                    count += 1
            if (count*1.0)/len(people) <= 0.5:
                movies.remove(mov)
        com_movies[i] = movies
#print 'mismatches: '+str(mismatches)

community_avgs = [[] for i in range(0, m)]
community_totals = [[] for i in range(0, m)]
for i in range(0, m):
    movies = com_movies[i]
    l = len(movies)
    if l < 5:
        community_avgs[i] = [i, 0,0,0,0]
        community_totals[i] = [i, 0,0,0,0]
    else:
        total_budget = 0
        total_rating = 0
        total_revenue = 0
        total_profit = 0
        for mov in movies:
            total_rating += movie_rating_dict[mov]
            total_budget += movie_budget_dict[mov]
            total_revenue += movie_revenue_dict[mov]
            total_profit += movie_profit_dict[mov]
        avg_budget = total_budget/l
        avg_rating = total_rating/l
        avg_revenue = total_revenue/l
        avg_profit = total_profit/l
        if total_budget == 0 or total_revenue == 0:
            community_avgs[i] = [i, 0,0,0,0]
            community_totals[i] = [i, 0,0,0,0]
        else:
            community_avgs[i] = [i, avg_rating, avg_budget, avg_revenue, avg_profit]
            community_totals[i] = [i,total_rating, total_budget, total_revenue, total_profit]

#cleans community list of communities that are too small

rm_c = []
rm_m = []
rm_a = []
for i in range(0, len(cluster)):
    if community_avgs[i][1] == 0:
        rm_c.append(cluster[i])
        rm_m.append(com_movies[i])
        rm_a.append(community_avgs[i])

for i in range(0,len(rm_c)):
    cluster.remove(rm_c[i])
    com_movies.remove(rm_m[i])
    community_avgs.remove(rm_a[i])

print len(cluster)
print len(com_movies)
print len(community_avgs)
# print com_movies[333]

for i in range(0, len(cluster)):
    if 'Samuel L Jackson' in cluster[i]:
        print 'slj' + str(i)
    if 'Seth Rogen' in cluster[i]:
        print 'sr' + str(i)
    if 'Jonah Hill' in cluster[i]:
        print 'jh' + str(i)
    if 'Ian McKellen' in cluster[i]:
        print 'im' + str(i)
    if 'Stellan Skarsgrd' in cluster[i]:
        print 'ss' + str(i)
    if 'Daniel Radcliffe' in cluster[i]:
        print 'DR' + str(i)
    if 'Leonardo DiCaprio' in cluster[i]:
        print 'Leo' + str(i)
```

```python
# find stdev and median of ratings and profits in each community

c_stdevs = []
c_medians = []

for i in range(0, len(cluster)):
    r_v = 0 #rating variance
    r_s = 0 #rating stdev
    p_v = 0 #profit var
    p_s = 0 #profit stdev
    avg_r = community_avgs[i][1]
    avg_p = community_avgs[i][4]
    movies = com_movies[i]
    plist = []
    rlist = []
    for mov in movies:
        new_r = movie_rating_dict[mov]
        new_p = movie_profit_dict[mov]
        r_v += (avg_r-new_r)**2
        p_v += (avg_p-new_p)**2
        rlist.append(new_r)
        plist.append(new_p)
    r_v = r_v/len(movies)
    p_v = p_v/len(movies)
    r_s = math.sqrt(r_v)
    p_s = math.sqrt(p_v)
    r_m = np.median(rlist)
    p_m = np.median(plist)
    c_medians.append((i, r_m, p_m))
    c_stdevs.append((i, r_s, p_s))

sorted_ratings = sorted(c_stdevs, key=lambda x: x[1], reverse=False)
print sorted_ratings[:10]
sorted_profits = sorted(c_stdevs, key=lambda x: x[2], reverse=False)
print sorted_profits[:10]
print
sorted_r_m = sorted(c_medians, key=lambda x: x[1], reverse=True)
print sorted_r_m[:10]
sorted_p_m = sorted(c_medians, key=lambda x: x[2], reverse=True)
print sorted_p_m[:10]

s_r = [x[1] for x in c_stdevs]
s_p = [x[2] for x in c_stdevs]
s_r_m = [x[1] for x in c_medians]
s_p_m = [x[2] for x in c_medians]

fig = plt.figure()
fig.suptitle('Communities: StDev (Rating vs. Profit)', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)

ax.set_xlabel('Rating')
ax.set_ylabel('Profit (in hundreds of millions)')
ax.scatter(s_r, s_p)

correlation = sp.stats.pearsonr(s_r,s_p)
print correlation

# print
# print cluster[157]
# print com_movies[157]
```
