# Clustering the data from the Mayor's 24-hour hotline in Boston

## Introduction 

In this assignment, we will do a clustering analysis on the 311 calls that we used in the previous homework. We will focus on choosing a value for k, clustering based on multiple features and detecting outliers.

---------------------------

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
boston = pd.read_csv('../Homework-2/cases_2015.csv')
new_column_names = [ x.lower().replace(' ','_') for x in boston.columns]
boston.columns = new_column_names
```

## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

For example, the result of a clustering with 10 clusters is the following:


![A clustering of the locations of the calls for k=10](clusters10.png)

Filter the dataset to get the latitude and longitude, transform it as required and run the k-means clustering algorithm. Argue about the number of clusters you used. **(pts 50)**

```python
import numpy as np
import scipy as sp
import sklearn.metrics as metrics
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
import seaborn as sns

kmeans = KMeans(init='k-means++', n_clusters=9, n_init=10)
bn = boston

geo_data = pd.DataFrame([bn.latitude, bn.longitude]).transpose()

scaled = StandardScaler().fit_transform(geo_data)
geo_data = pd.DataFrame(scaled, columns=['latitude', 'longitude'])

kmeans.fit_predict(geo_data)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

# print "The total error of the clustering is: ", error
# print '\nCluster labels'
# print labels
# print '\n Cluster Centroids'
# print centroids

    
plt.scatter(geo_data.longitude, geo_data.latitude, c=labels, cmap=plt.cm.Set1, marker='+',s=15, alpha=0.8)
```

I used nine clusters, because the jump from nine to ten clusters was the first that did not yield a net loss of more than 1000 error.  Nine clusters looked like the knee of the clustering graph.

## Step 2. Using multiple features

The next step in our analysis is to use more features than just the geolocation of a record (whose clustering only gives us information about the neighborhoods, which should not be a huge revelation!) Filter the dataset to get more features, transform them as required and compute another clustering. You should argue about the quality of your clustering and use domain knowledge (i.e. the fact that you are locals) to explore the clusters you produced and present them to the reader. **(pts 30)**

```python
bn = boston

bn = bn.fillna(0)

date_data = pd.DataFrame([bn.open_dt]).transpose()
date_data = date_data.open_dt.map(lambda x: x[:10])

date_encoder = LabelEncoder()
date_data = date_encoder.fit_transform(date_data)

title_data = pd.DataFrame([bn.case_title]).transpose()

title_encoder = LabelEncoder()
title_data = title_encoder.fit_transform(title_data)

zip_data = pd.DataFrame([bn.location_zipcode]).transpose()

zip_encoder = LabelEncoder()
zip_data = zip_encoder.fit_transform(zip_data)

time_data = pd.DataFrame([bn.ontime_status]).transpose()
ontime_encoder = LabelEncoder()
time_data = ontime_encoder.fit_transform(time_data)

new_data = pd.DataFrame([date_data, title_data, zip_data, time_data]).transpose()
new_data.columns = ['date', 'title', 'zip', 'ontime']
#print new_data

geo_data = pd.DataFrame([bn.latitude, bn.longitude]).transpose()

scaled = StandardScaler().fit_transform(geo_data)
geo_data = pd.DataFrame(scaled, columns=['latitude', 'longitude'])

final_data = pd.DataFrame([geo_data.latitude, geo_data.longitude, new_data.date, new_data.title, new_data.zip, new_data.ontime])

final_data = normalize(final_data, norm='l2', axis=1, copy=True)

final_data = pd.DataFrame([final_data[0], final_data[1], final_data[2], final_data[3], final_data[4], final_data[5]]).transpose()
final_data.columns = geo_data.columns + new_data.columns
final_data = final_data*100
#print final_data

kmeans = KMeans(init='k-means++', n_clusters=6, n_init=15)

kmeans.fit_predict(final_data)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

print "The total error of the clustering is: ", error
print '\nCluster labels'
print labels
print '\n Cluster Centroids'
print centroids

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,15))
#ax.scatter(new_data.zip, new_data.ontime, c=labels, cmap=plt.cm.Set1, marker='+',s=15, alpha=0.8)
ax1.scatter(final_data.longitude, final_data.latitude, c=labels, cmap=plt.cm.Set1, marker='+',s=15, alpha=0.8)

distances = metrics.euclidean_distances(final_data.head(1000))
idx = np.argsort(labels[:1000])
re_di = distances[idx,:][:,idx]
sns.heatmap(re_di, xticklabels=False, yticklabels=False, linewidths=0, square=True,cbar=False)
```
